```python
# import library yang dibutuhkan
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np

# membuat variabel untuk membaca gambar dengan warna yang dibaca keabu abuan
citra1 = cv.imread("rahmat.jpg" , cv.IMREAD_GRAYSCALE)
citra2 = cv.imread("rahmat2.jpg", cv.IMREAD_GRAYSCALE)

# mencetak ukuran citra 1 dan 2
print('Shape citra 1 : ', citra1.shape)
print('Shape citra 1 : ', citra2.shape)

# mencetak gambar citra dalam 1  figure
# 1 baris 2 kolom 10x10
fig, axes = plt.subplots(1, 2, figsize=(10, 10))
ax = axes.ravel()

ax[0].imshow(citra1, cmap = 'gray')
ax[0].set_title("Citra 1")
ax[1].imshow(citra2, cmap = 'gray')
ax[1].set_title("Citra 2")
```

```python
# membuat ukuran baris kolom citra
# membuat matriks kosong 
m1,n1 = citra1.shape
output1 = np.empty([m1, n1])
m2,n2 = citra2.shape
output2 = np.empty([m2, n2])

# mencetak ukuran citra 
print('Shape copy citra 1 : ', citra1.shape)
print('Shape output citra 1 : ', output1.shape)

print('m1 : ',m1)
print('n1 : ',n1)
print()

print('Shape copy citra 2 : ', citra2.shape)
print('Shape output citra 3 : ', output2.shape)
print('m2 : ',m2)
print('n2 : ',n2)
print()
```

```python
# Proses Filter Median Pada Citra Input 1
for baris in range(0, m1-1):
    for kolom in range(0, n1-1):
        a1 = baris
        b1 = kolom
        dataA = [citra1[a1-1, b1-1], citra1[a1-1, b1], citra1[a1-1, b1+1], \
              citra1[a1, b1-1], citra1[a1, b1], citra1[a1, b1+1], \
              citra1[a1+1, b1-1], citra1[a1+1, b1], citra1[a1+1, b1+1]]
        
        # Urutkan
        for i in range(1, 8):
            for j in range(i, 9):
                if dataA[i] > dataA[j]:
                    tmpA = dataA[i];
                    dataA[i] = dataA[j];
                    dataA[j]= tmpA;
        
        output1[a1, b1] = dataA[5]
```

```python
# Proses Filter Median Pada Citra Input 2
for baris in range(0, m2-1):
            for kolom in range(0, n2-1):
                a1 = baris
                b1 = kolom
                dataA = [citra2[a1-1, b1-1], citra2[a1-1, b1], citra2[a1-1, b1+1], \
                    citra2[a1, b1-1], citra2[a1, b1], citra2[a1, b1+1], \
                    citra2[a1+1, b1-1], citra2[a1+1, b1], citra2[a1+1, b1+1]]
                
                # Urutkan
                for i in range(1, 8):
                    for j in range(i, 9):
                        if dataA[i] > dataA[j]:
                            tmpA = dataA[i];
                            dataA[i] = dataA[j];
                            dataA[j]= tmpA;
        
output2[a1, b1] = dataA[5]
```

```python
# Plot Gambar
#  2 baris 2 kolom 10x10
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
ax = axes.ravel()

ax[0].imshow(citra1, cmap = 'gray')
ax[0].set_title("Input Citra 1")

ax[1].imshow(citra2, cmap = 'gray')
ax[1].set_title("Input Citra 1")

ax[2].imshow(output1, cmap = 'gray')
ax[2].set_title("Output Citra 1")

ax[3].imshow(output2, cmap = 'gray')
ax[3].set_title("Output Citra 2")
```
