```python
# import library yang dibutuhkan
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np

# membuat variabel untuk membaca gambar dengan warna yang dibaca keabu abuan
citra1 = cv.imread("rahmat.jpg" , cv.IMREAD_GRAYSCALE)
citra2 = cv.imread("rahmat2.jpg", cv.IMREAD_GRAYSCALE)

# mencetak ukuran citra 1 dan 2
print('Shape citra 1 : ', citra1.shape)
print('Shape citra 1 : ', citra2.shape)

# mencetak gambar citra dalam 1  figure
# 1 baris 2 kolom 10x10
fig, axes = plt.subplots(1, 2, figsize=(10, 10))
ax = axes.ravel()

ax[0].imshow(citra1, cmap = 'gray')
ax[0].set_title("Citra 1")
ax[1].imshow(citra2, cmap = 'gray')
ax[1].set_title("Citra 2")
```

```python
# membuat ukuran baris kolom citra
# membuat matriks kosong 
m1,n1 = citra1.shape
output1 = np.empty([m1, n1])
m2,n2 = citra2.shape
output2 = np.empty([m2, n2])

print('Shape copy citra 1 : ', citra1.shape)
print('Shape output citra 1 : ', output1.shape)

print('m1 : ',m1)
print('n1 : ',n1)
print()

print('Shape copy citra 2 : ', citra2.shape)
print('Shape output citra 3 : ', output2.shape)
print('m2 : ',m2)
print('n2 : ',n2)
print()
```

```python
# Proses Filter Batas Pada Citra Input 1
for baris in range(0, m1-1):
    for kolom in range(0, n1-1):
        
        a1 = baris
        b1 = kolom
        
        arr = np.array([citra1[a1-1, b1-1], citra1[a1-1, b1], citra1[a1, b1+1], \
            citra1[a1, b1-1], citra1[a1, b1+1], citra1[a1+1, b1-1],  \
            citra1[a1+1, b1], citra1[a1+1, b1+1]])
        
        minPiksel = np.amin(arr);        
        maksPiksel = np.amax(arr);    
            
        if citra1[baris, kolom] < minPiksel :
            output1[baris, kolom] = minPiksel
        else :
            if citra1[baris, kolom] > maksPiksel :
                output1[baris, kolom] = maksPiksel
            else :
                output1[baris, kolom] = citra1[baris, kolom]
```

```python
# Proses Filter Batas Pada Citra Input 2
for baris1 in range(0, m2-1):
    for kolom1 in range(0, n2-1):
        
        a1 = baris1
        b1 = kolom1
        
        arr = np.array([citra2[a1-1, b1-1], citra2[a1-1, b1], citra2[a1, b1+1], \
            citra2[a1, b1-1], citra2[a1, b1+1], citra2[a1+1, b1-1],  \
            citra2[a1+1, b1], citra2[a1+1, b1+1]])
        
        minPiksel = np.amin(arr);        
        maksPiksel = np.amax(arr);    
            
        if citra2[baris1, kolom1] < minPiksel :
            output2[baris1, kolom1] = minPiksel
        else :
            if citra2[baris1, kolom1] > maksPiksel :
                output2[baris1, kolom1] = maksPiksel
            else :
                output2[baris1, kolom1] = citra2[baris1, kolom1]
```

```python
# Plot Citra Input dan Output Hasil dari Filter Batas
# memasukan gambar dalam 1 figure dengan 2 baris 2 kolom berukuran 10x10
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
ax = axes.ravel()

ax[0].imshow(citra1, cmap = 'gray')
ax[0].set_title("Input Citra 1")

ax[1].imshow(citra2, cmap = 'gray')
ax[1].set_title("Input Citra 2")

ax[2].imshow(output1, cmap = 'gray')
ax[2].set_title("Output Citra 1")

ax[3].imshow(output2, cmap = 'gray')
ax[3].set_title("Output Citra 2")
```
