```python
#import library yang dibutuhkan
import matplotlib.pyplot as plt
from skimage.io import imread
import numpy as np
import cv2

# membuat variabel untuk membaca file gambar dan mencetak ukurannya
citra1 = imread("rahmat.jpg")
print(citra1.shape)

#mencetak gambar
plt.imshow(citra1, cmap='gray')
```

```python
# Proses Konvolusi
# Membuat matriks
kernel = np.array([[-1, 0, -1], 
                   [0, 4, 0], 
                   [-1, 0, -1]])

citraOutput = cv2.filter2D(citra1, -1, kernel) # melakukan konvolusi antara citra 1 dengan kernel , dengan depth -1 yg menghasilkan depth yang sama antara output dan input

fig, axes = plt.subplots(1, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(citra1, cmap = 'gray')
ax[0].set_title("Citra Input")
ax[1].imshow(citraOutput, cmap = 'gray')
ax[1].set_title("Citra Output")
```
