```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#membuat variabel img untuk membaca file rahmat.jpg dengan warna grayscale
img = cv2.imread("rahmat.jpg", cv2.IMREAD_GRAYSCALE)

#membuat dimensi dari gambar
row, column = img.shape

#membuat array kosong untuk dimensi gambar
img1 = np.zeros((row,column),dtype = 'uint8')
 
# membuat batas minimum dan  maksimum
min_range = 10
max_range = 60
 
# fungsi membuat grayscale menjadi hitam putih
for i in range(row): # melakukan perulangan untuk setiap baris pada gambar
    for j in range(column): # melakukan perulangan untuk setiap  kolom pada gambar
        if img[i,j]>min_range and img[i,j]<max_range: # membuat aturan jika nilai sebuah pixel melebihi batas minimum dan kurang dari batas maksimum baris dan kolom berisi 255
            img1[i,j] = 255
        else: # selain dari dua kemungkinan tadi maka nilainya 0
            img1[i,j] = 0

# Menampilkan hasil semua dalam 1 figure 2 baris 2 kolom berukuran 12x12
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

# baris 1 kolom 1 dan 2
ax[0].imshow(img, cmap=plt.cm.gray)# menampilkan gambar
ax[0].set_title("Citra Input") # judul gambar
ax[1].hist(img.ravel(), bins=256) # menampilkan histogram gambar dengan intensitas maksimal 256
ax[1].set_title('Histogram Input')

# baris 2 kolom 1 dan 2
ax[2].imshow(img1, cmap=plt.cm.gray)
ax[2].set_title("Citra Output")
ax[3].hist(img1.ravel(), bins=256)
ax[3].set_title('Histogram Output')

fig.tight_layout() # membuat padding pada suplot agar lebih rapih
```
