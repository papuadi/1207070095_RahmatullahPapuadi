```python
# Untuk brightness maka ditambah konstata akan menambah seluruh array pixel
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread("rahmat2.jpg")

# Membuat variabel tinggi, lebar dan channel untuk citra
img_height = img.shape[0]
img_width = img.shape[1]
img_channel = img.shape[2]
img_type = img.dtype

# Membuat array kosong seukuran citra
img_brightness = np.zeros(img.shape, dtype=np.uint8)

# Membuat fungsi dengan def
# fungsi untuk mengatur tingkat kecerahan
def brighter(nilai):
    for y in range(0, img_height):
        for x in range(0,img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red)+int(green)+int(blue))/3
            gray += nilai
            if gray > 255 :
                gray = 255
            if gray < 0 :
                gray = 0
            img_brightness[y][x] =  (gray,gray,gray)

# memanggil fungsi untuk mengatur kecerahan untuk citra
brighter(-50)
plt.imshow(img_brightness)
plt.title("Brightness -50")
plt.show() 

brighter(+50)
plt.imshow(img_brightness)
plt.title("Brightness +50") 

plt.show()

# Membuat fungsi untuk mengatur nilai RGB
img_rgbbright = np.zeros(img.shape, dtype=np.uint8)
def rgbbrighter(nilai):
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            if red > 255 :
                red = 25
            if red < 0 :
                red = 0
            green = img[y][x][1]
            green += nilai
            if green > 255 :
                green = 255
            if green < 0 :
                green = 0
            blue = img[y][x][2]
            blue += nilai
            if blue > 255:
                blue = 255
            if blue < 0 :
                blue = 0
            img_rgbbright[y][x] = (red,green,blue)
        
rgbbrighter(-100)
plt.imshow(img_rgbbright)
plt.title("rgb -100")
plt.show() 

rgbbrighter(+100)
plt.imshow(img_rgbbright)
plt.title("rgb +100") 

plt.show()


# membuat fungsi contrast
img_contrass = np.zeros(img.shape, dtype= np.uint8)

def contrass(nilai):
    for y in range(0,img_height):
        for x in range(0,img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red)+int(green)+int(blue))
            if gray > 255:
                gray = 255
            img_contrass[y][x] = (gray, gray, gray)
            
# Menampilkan
contrass(2)
plt.imshow(img_contrass)
plt.title("contrast 2")
plt.show() 

contrass(4)
plt.imshow(img_contrass)
plt.title("contranst 4") 

plt.show()


# membuat fungsi rgb contrast
img_rgbcontrass = np.zeros(img.shape, dtype=np.uint8)

def rgbcontrass(nilai):
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            if red > 255 :
                red = 25
            green = img[y][x][1]
            green += nilai
            if green > 255 :
                green = 255
            blue = img[y][x ][2]
            blue += nilai
            if blue > 255:
                blue = 255
            img_rgbcontrass[y][x] = (red,green,blue)
            
# Menampilkan
rgbcontrass(2)
plt.imshow(img_rgbcontrass)
plt.title("rgb contrast 2")
plt.show() 

rgbcontrass(3)
plt.imshow(img_rgbcontrass)
plt.title("rgb contranst 3") 
plt.show()

# Membuat variabel img_contrass untuk menampung hasil
img_autocontrass = np.zeros(img.shape, dtype=np.uint8)
# Melakukan penambahan contrass secara otomatis
def autocontrass():
    xmax = 300
    xmin = 0
    d = 0
    # Mendapatkan nilai d, dimana nilai d ini akan berpengaruh pada hitungan
    # untuk mendapatkan tingkat kontras
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            if gray < xmax:
                xmax = gray
            if gray > xmin:
                xmin = gray
    d = xmin-xmax
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            gray = int(float(255/d) * (gray-xmax))
            img_autocontrass[y][x] = (gray, gray, gray)
#  Menampilkan hasil autolevel contrass
autocontrass()
plt.imshow(img_autocontrass)
plt.title("Contrass Autolevel")
plt.show()

```
