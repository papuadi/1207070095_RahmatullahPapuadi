```python
import matplotlib.pyplot as plt
%matplotlib inline

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray 
from skimage.util import invert

import numpy as np
# Percobaan 1 - Cropping Image

# membaca file citra dan membuatnya dengan variabel
img1 = imread("rahmat.jpg")
img2 = imread("rahmat2.jpg")

# membuat variabel citra namun citra asli dipotong
img1crop = img1.copy()
img1crop = img1crop[0:256,64:320] # letak pemotongan sb x , sb y

# membuat variabel citra namun citra asli dipotong
img2crop = img2.copy()
img2crop = img2crop[64:256,128:320] # letak pemotongan sb x , sb y

# mencetak ukuran img 1 dan img 2 
print('Ukuran img 1 : ',img1.shape)
print('Ukuran img 1 crop : ',img1crop.shape)
print('Ukuran img 2 : ',img2.shape)
print('Ukuran img 2 crop : ',img2crop.shape)

# membuat figure dengan 2 baris 2 kolom dengan ukuran 12x12
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

# posisi baris 1 kolom 1
ax[0].imshow(img1)
ax[0].set_title("Citra Input 1") # menampilkan img1
# posisi baris 1 kolom 2
ax[1].imshow(img2, cmap='gray') # menampilkan img2
ax[1].set_title('Citra Input 2')
# posisi baris 2 kolom 1
ax[2].imshow(img1crop)
ax[2].set_title("Citra Output 1") # menampilkan img1crop
# posisi baris 2 kolom 2
ax[3].imshow(img2crop, cmap='gray') # menampilkan img2crop
ax[3].set_title('Citra Output 2')


# Percobaan 2 - Citra Negative

#membuat img1crop berwarna negatif
inv = invert(img1crop)

#mencetak ukuran img1crop dan img1crop negatif
print('Shape Input : ', img1crop.shape)
print('Shape Output : ',inv.shape)

# membuat figure dengan 2 baris 2 kolom dengan ukuran 12x12
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(img1crop)
ax[0].set_title("Citra Input")

ax[1].hist(img1crop.ravel(), bins=256) # menampilkan histogram img1crop
ax[1].set_title('Histogram Input')

ax[2].imshow(inv)
ax[2].set_title('Citra Output (Inverted Image)')

ax[3].hist(inv.ravel(), bins=256)  # menampilkan histogram img1crop negatif
ax[3].set_title('Histogram Output')

# membuat variabel  img2crop2 dalam bentuk array
img2crop2 = np.array(img2crop, dtype=float)

for baris in range(0, img2crop2.shape[0]-1):
    for kolom in range(0, img2crop2.shape[1]-1):
        img2crop2[baris, kolom] + 100

# membuat figure dengan 2 baris 2 kolom dengan ukuran 12x12
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(img2crop, cmap='gray')
ax[0].set_title("Citra Input")

ax[1].hist(img2crop2.ravel(), bins=256)
ax[1].set_title('Histogram Input')

ax[2].imshow(img2crop2, cmap='gray')
ax[2].set_title('Citra Output (Brightnes)')

ax[3].hist(img2crop2.ravel(), bins=256)
ax[3].set_title('Histogram Output')
```
